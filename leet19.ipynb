```python
arr = [2, 8, 1, 3, 6, 7, 5, 4]

def quicksort(arr):
    # print(arr)
    def partition(arr):
        pivot = len(arr)-1
        part_idx = -1
        i = 0
        # print(arr[pivot])

        while i < len(arr[:pivot]):
            # print(arr[i], arr[pivot], part_idx)
            if arr[i] > arr[pivot] and part_idx < 0:
                part_idx = i
                # print(arr[i], arr[part_idx])
            elif arr[i] <= arr[pivot] and part_idx >= 0:
                arr[i], arr[part_idx] = arr[part_idx], arr[i]
                part_idx += 1
            i += 1
        arr[part_idx], arr[pivot] = arr[pivot], arr[part_idx]

        return part_idx
    
    if len(arr) <= 1:
        return arr
    
    part_idx = partition(arr)

    return quicksort(arr[:part_idx]) + quicksort(arr[part_idx:])
    # return arr[:part_idx] + arr[part_idx:]
    # return quicksort(arr[:partition(arr)]) + quicksort(arr[partition(arr):])

quicksort(arr)
```
